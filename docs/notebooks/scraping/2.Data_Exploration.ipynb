```python
import pandas as pd
import glob

from bokeh.palettes import Spectral4
from bokeh.plotting import figure, show
from bokeh.palettes import Category10, Category20, Category20b, Category20c
from bokeh.palettes import Greys256, Inferno256, Magma256, Plasma256
```

# Loading the data

```python
#load the datasets and combine

extension = 'csv'
all_filenames = [i for i in glob.glob('scraped_data/*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])

#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False)
```

# Analysis

```python
combined_csv.shape
```

```python
combined_csv.columns
```

```python
items_list = combined_csv['name'].unique()
items_list = ['Zuchini Green (Price per 500 gms)',
              'Zuchini Yellow (Price per 250gms)']
```

```python
p = figure(plot_width=1400, plot_height=600, x_axis_type="datetime")
p.title.text = 'Click on legend entries to mute the corresponding lines'

for name, color in zip(items_list, Inferno256):
    df = combined_csv[combined_csv['name']==name].reset_index(drop=True)
    df.sort_values(by='date',inplace=True)
    df['date'] = pd.to_datetime(df['date'])
    p.line(df['date'], df['price'], line_width=2, color=color, alpha=0.8,
           muted_color=color, muted_alpha=0.2, legend_label=name)

p.legend.location = "bottom_left"
p.legend.click_policy="mute"

show(p)
```
